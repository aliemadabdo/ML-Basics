```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
from PIL import Image
```

```python
x = np.linspace(-10, 10, 1000)

def seg(net):
  return 1/(1+math.e**(-net))
def tanh(net):
  return ( np.exp(net) - np.exp(-net) ) / ( np.exp(net) + np.exp(-net) )
def relu(net):
    return np.maximum(0, net)
def drelu (net):
  return np.logical_xor(0,net)



dseg = seg(x)*(1-seg(x))
dtanh = 1 - tanh(x)**2
```

```python
plt.plot(x,seg(x)) 
plt.plot(x,dseg) 
```

```python
plt.plot(x,tanh(x)) 
plt.plot(x,dtanh) 
```

```python
plt.plot(x,relu(x)) 
plt.plot(x,drelu(relu(x))) 
```
