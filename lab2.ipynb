```python
import matplotlib.pyplot as plt
import numpy as np

def read_image(filename):
    return plt.imread(filename)

def plot_image(image):
    plt.imshow(image)
    plt.show()

def split_components(image):
    return image[:,:, 0], image[:,:, 1], image[:,:, 2]

def plot_component(component, title):
    component_image = image.copy()
    
    rgbDict={'r':0 ,'g':1 ,'b':2}
    component_image[:, :, rgbDict[component]] = 0
    component_image[:, :, 2] = 0
    plt.imshow(component, cmap='gray')
    plt.title(title)
    plt.show()
    component_image = image.copy()


def max_gray(image):
    return np.max(image, axis=2)

def avg_gray(image):
    return np.mean(image, axis=2)

def eq_gray(image):
    return 0.2126 * image[:,:, 0] + 0.7152 * image[:,:, 1] + 0.0722 * image[:,:, 2]

image = read_image("example.jpg")
plot_image(image)
r, g, b = split_components(image)
plot_component('r', "Red")
plot_component('g', "Green")
plot_component('b', "Blue")

gray_max = max_gray(image)
plot_component(gray_max, "Max")
gray_avg = avg_gray(image)
plot_component(gray_avg, "Average")
gray_eq = eq_gray(image)
plot_component(gray_eq, "Equation")
```
